```python
import pandas as pd
import numpy as np
```

```python
customers = pd.read_csv("olist_customers_dataset.csv")
```

```python
geolocation = pd.read_csv("olist_geolocation_dataset.csv")
```

```python
order_items = pd.read_csv("olist_order_items_dataset.csv")
```

```python
payments = pd.read_csv("olist_order_payments_dataset.csv")
```

```python
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
print(orders.head())
print(orders.info())
print(orders.isnull().sum())
```

```python
date_columns = [
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])
print(orders)       
```

```python
products =  pd.read_csv("olist_products_dataset.csv")
```

```python
sellers = pd.read_csv("olist_sellers_dataset.csv")
```

```python
category = pd.read_csv("product_category_name_translation.csv")
```

```python
df = orders.merge(order_items, on="order_id", how="inner")
print(df)
```

```python
df = df.merge(products, on="product_id", how="left")
print(df)
```

```python
df = df.merge(customers, on="customer_id", how="left")
print(df)
```

```python
df = df.merge(payments, on="order_id", how="left")
print(df)
```

```python
df["total_item_value"] = df["price"] + df["freight_value"]
print(df)
```

```python
df["order_month"] = df["order_purchase_timestamp"].dt.to_period("M")
print(df)
```

```python
monthly_revenue = df.groupby("order_month")["total_item_value"].sum()
print(monthly_revenue)
```

```python
top_categories = (
    df.groupby("product_category_name")["total_item_value"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)
print(top_categories)
```

```python
state_revenue = (
    df.groupby("customer_state")["total_item_value"]
      .sum()
      .sort_values(ascending=False)
)
print(state_revenue)
```

```python
customer_spend = df.groupby("customer_id")["total_item_value"].sum()
mean_spend = np.mean(customer_spend)
median_spend = np.median(customer_spend)
percentile_90 = np.percentile(customer_spend, 90)
print(customer_spend)
print(mean_spend)
print(median_spend)
print(percentile_90)
```

```python
monthly_revenue.to_csv("monthly_revenue.csv")
top_categories.to_csv("top_categories.csv")
state_revenue.to_csv("state_revenue.csv")
```
